```python
# Before I begin I'll do some more veeeeery necessary spring cleaning
# Clean and convert salary column properly, this was causing alot of bugs within my code since this column
# contained a range of values that had string characters, affecting type conversions and int based functions

def clean_salary(salary_str):
    """Clean salary string and extract numeric value"""
    if pd.isna(salary_str): # Null-guard:  If the value is empty, null, None or any of these other values return immediately
        return np.nan
    
    # Convert to string and clean
    salary_str = str(salary_str).strip() 
    
    # Remove dollar signs, commas, and other common formatting
    salary_str = salary_str.replace('$', '').replace(',', '')
    
    # Next Handle ranges by taking the first number (we'll take the conservative approach)
    if '-' in salary_str:
        salary_str = salary_str.split('-')[0].strip()
    
    # Remove any remaining non-numeric characters except decimal points
    import re
    salary_str = re.sub(r'[^\d.]', '', salary_str)
    
    # Return cleaned string for pd.to_numeric to handle
    return salary_str if salary_str else np.nan

# Apply cleaning function then convert to numeric
df_salary['salary_cleaned'] = df_salary['salary'].apply(clean_salary)
df_salary['salary'] = pd.to_numeric(df_salary['salary_cleaned'], errors='coerce')

# Drop the temporary cleaning column
df_salary = df_salary.drop('salary_cleaned', axis=1)
df_salary.head(10)
```

```python
# A couple columns have incorrect datatypes, lets adjust accordingly
# First, timestamp column
df_salary['timestamp'] = pd.to_datetime(df_salary['timestamp'], errors='coerce')

# Next standardize US country names
df_salary['country'] = df_salary['country'].apply(standardize_us_names)

df_salary.head(10)
```

```python
# Create a copy for the new test data
salary_test = df_salary.copy()

# I'll create some salary statistics by state first (similar to what I did before)
# Filter for USD and clean country names
cleaned_salary_df = salary_test[salary_test['currency'] == 'USD'].copy()
cleaned_salary_df = cleaned_salary_df[cleaned_salary_df['country'] == 'US']

# Then create salary statistics by state
salary_stats = (
    cleaned_salary_df.groupby('state')['salary']
    .agg(['min', 'mean', 'max', 'count'])
    .sort_values('mean', ascending=False)
)

print(salary_stats)
```

### Bonus #1

This time, format the table output nicely (*$12,345.00*) without modifying the values in the `DataFrame`. That is, `df_salary` should be identical before versus after running your code.

(*Hint: if you run into an error about `jinja2` perhaps you need to `pip install` something.*)

```python
# So I'll create a function for formatting the salary column
def currency_formatter(salary):
    if pd.isna(salary):
        return ""
    return "${:,.2f}".format(salary)

# Then apply formatting to the salary statistics but create a fallback for potential issues
# like uninstalled dependencies or missing data

try:
    formatted_salary_stats = salary_stats.style.format({
        'min': currency_formatter,
        'mean': currency_formatter, 
        'max': currency_formatter,
        'count': '{:.0f}'  # Count doesn't need currency formatting
    })
    
    print("\nFormatted salary statistics:")
    display(formatted_salary_stats)
    
except Exception as e:
    print(f"Styling error: {e}")
    print("Falling back to manual formatting...")
    
    # Alternative formatted copy for display
    formatted_display = salary_stats.copy()
    for col in ['min', 'mean', 'max']:
        formatted_display[col] = formatted_display[col].apply(currency_formatter)
    
    print(formatted_display)
```

### Bonus #3

Show the quantiles instead of just minimum, mean, and maximumâ€•say 0%, 5%, 25%, 50%, 75%, 95%, and 100%. Outliers may be deceiving.

Sort by whatever interests youâ€•like say the *50th* percentile.

And throw in a count by state too. It would be interesting to know how many data points contribute to the figures for each state. (*Hint: your nice formatting from Bonus #1 might not work this time around.* ðŸ˜œ)

```python
# So the question is asking me to perform a statistical analysis to find and show the different quantiles 
# Since we'll be doing a count by state, I figure we'll be working with data filtered for the US specifically,
# I'll be working with the single_state_df I created in the previous bonus question

def calculate_quantiles(salary_series):
    """Function to calculate quantiles: 0%, 5%, 25%, 50%, 75%, 95% and 100% along with the count"""
    quantiles = [0.0, 0.05, 0.25, 0.50, 0.75, 0.95, 1.00]
    res = {}

    for q in quantiles:
        res[f'{int(q*100)}%'] = salary_series.quantile(q)

    # Also add the count of data points
    res['count'] = len(salary_series)

    return pd.Series(res)

# I'll groupby State and calculate the quantiles, then sort by the 50th percentile (or the median)
salary_quantiles_by_state = (single_state_df.groupby('state')['salary']
        .apply(calculate_quantiles).unstack().sort_values('50%', ascending=False))

salary_quantiles_by_state
```
